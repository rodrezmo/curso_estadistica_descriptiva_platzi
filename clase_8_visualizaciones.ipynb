```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
print(iris['species'].nunique(),',',iris['species'].unique())
```

Vamos a utilizar el diagrama de dispersión para tener un analisis de mas de una variable, podemos ver la evolución de ambas a la vez y determinar si estan o no correlacionadas. Tambien podemos ver si esa correlación es inversa o directa. 

Si dibujan una recta diagonal donde ambas crecen diremos que estan directamente relacionadas.

Si dibujan una recta diagonal donde una crece y la otra decrece estan inversamente relacionadas. 

Si no dibujan una recta diagonal podemos decir que no estan correlacionados.

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length', hue='species')
```

Vemos que el largo del petalo y sepalo no estan 100% relacionadas ya que la setosa no presenta una correlación tan positiva como si las otras 2 species.

```python
sns.jointplot(data=iris,x='petal_width',y='petal_length',palette='inferno',hue='species')
```

Mientras que si comparamos los atributos del petalo, vemos que todas las species tienen alta correlación

```python
sns.jointplot(data=iris,y='sepal_width',x='sepal_length',palette='inferno',hue='species')
```

Mientras que para los atributos del sepalo vemos que los datos no tienen correlación ya que estan muy dispersos y no forman una linea recta diagonal

```python
sns.boxplot(data=iris,x='species',y='sepal_width')
```

```python
iris_agg = iris.groupby('species')['petal_width'].mean().reset_index()
```

```python
sns.barplot(data=iris_agg,x='species',y='petal_width')
```
