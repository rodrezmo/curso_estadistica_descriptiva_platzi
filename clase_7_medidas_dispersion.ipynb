Rango: definido por los valores minimo y maximo del conjunto.

Los Q, son particiones del conjunto con la misma cantidad de elementos. El Q2 es la mitad del conjunto, igual que la mediana.

el Rango inter quartil (IQR) se da por los valores de Q1 y Q3.

IQR = Q3 - Q1

DESVIACION ESTANDAR

Es el promedio de cuanto hay de distancia entre los valores y la media. En una distribución normal de datos se puede usar para determinar outliers

min = Q1 - 1.5IQR

max = Q3 + 1.5IQR

Lo que este fuera de esos rangos podemos considerarlos outliers, con el 1.5IQR es como si usaras 3 desviaciones estandar. Se abarca aprox. el 99.72% de los datos.

Para datos que no esten normalmente distrbuidos se utiliza otra fórmula

min = Q1 - f(IQR)

max = Q3 + g(IQR)

donde f y g son funciones en función del IQR, las mismas se usan para denotar el sesgo de la asimetría en la distribución.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
```

```python
# Desviacion estandar

df['price_usd'].std()
```

```python
# Rangos = valor max - valor min
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
# Quartiles
median = df['price_usd'].median()
q1 = df['price_usd'].quantile(q=0.25)
q3 = df['price_usd'].quantile(q=0.75)
min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1)

print(min_val,q1,median,q3,max_val)
```

```python
#iqr
iqr = q3 - q1
iqr
```

Límites para detección de outliers (datos simétricamente distribuidos)

datos entre Q1 - 1.5 x IQR y Q3 + 1.5 x IQR

```python
minlimit = q1 - 1.5 * iqr
maxlimit = q3 + 1.5 * iqr
print(minlimit,maxlimit)
```

El valor minimo del limite marca negativo ya que los datos no presentan una distribución normal y estan 'corridos' hacia la derecha
